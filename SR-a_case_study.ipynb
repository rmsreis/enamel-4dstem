```python
filename = 'Bin(2,2,2, 2) of outer enamel.dm4'
datacube = py4DSTEM.file.io.read(filename)
```

```python
filename = 'Diffraction of vacuum_256x256_50_25.dm3'
from ncempy.io import dm
vacuum = dm.dmReader(filename)
```

```python
#normalize by vacuum probe
vacuum_sum=np.sum(vacuum['data'])
normalized_data=datacube.data/vacuum_sum
```

```python
pixel_rad = 53.77034498680476
peak_max=np.zeros((cropped_image.shape[0],cropped_image.shape[1]))
remain=np.zeros((cropped_image.shape[0],cropped_image.shape[1]))

n_dots = 180   # set number of dots
angs = np.linspace(0, 2*np.pi, n_dots)  # angles to the dots
cx, cy = (jc, ic)  # center of circle
ra = pixel_rad 


for i in range(0,cropped_image.shape[0]):
        for j in range(0,cropped_image.shape[1]): 
            #image=np.average(datacube.data[i-3:i+3,j-3:j+3,:,:], axis=(0,1))
            image=cropped_image[i,j,:,:]
            xs, ys = [], []    
      

            for ang in angs:
                x = cx + ra*np.cos(ang)
                y = cy + ra*np.sin(ang)
                xs.append(x)   
                ys.append(y)   

            xs=np.asarray(xs)
            ys=np.asarray(ys)

            zi=np.zeros((1,len(xs)))
            for ii in range(0,len(xs)):
                    zi[0,ii] = np.average(image[xs[ii].astype(np.int)-3:xs[ii].astype(np.int)+3, \
                               ys[ii].astype(np.int)-3:ys[ii].astype(np.int)+3])
            
            max_loc=np.argmax(zi)
            peak_max[i,j]=max_loc
            remain[i,j]=max_loc%30 
```
